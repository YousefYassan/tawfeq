```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
#sklearn
from sklearn.preprocessing import OneHotEncoder,StandardScaler 
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.svm import SVR
```

```python
file_path = os.path.join(os.getcwd(),'PDataConcatData.csv')
df = pd.read_csv(file_path)
df
```

```python
df.isna().sum()
```

```python
df.dtypes
```

```python
df.describe()
```

```python
# Split X & y
X = df.drop(columns=['title','price before promotion','price',"website","currancy","rate",'img','url','reviews'], axis=1)
y = df['price']

```

```python
X_train, X_test , y_train , y_test=train_test_split(X,y,shuffle=True,test_size=0.30,random_state=42)
```

```python
print("X_train shape",X_train.shape)
print("t_train shape",y_train.shape)
print("X_test shape",X_test.shape)
print("y_test shape",y_test.shape)
```

```python
num_cols = [col for col in X_train.columns if X_train[col].dtype in ['int32','int64','float32','float64']]
categ_cols = [col for col in X_train.columns if X_train[col].dtype not in ['int32','int64','float32','float64']]
categ_cols,num_cols
```

```python
X.dtypes
```

```python
# data nemric pipline will scale it
num_pipline = Pipeline(steps = [
                ('imputer', SimpleImputer(strategy='median')),
                ('scaler',StandardScaler())
                ])
# data catg (str) will transform 
categ_pipline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy='constant', fill_value='missing')),
    ('ohe', OneHotEncoder(sparse_output=False, handle_unknown='ignore'))
])
#  concat catg with numric
all_pipeline = ColumnTransformer(transformers=[
                    ('numerical', num_pipline, num_cols),
                    ('categorical', categ_pipline, categ_cols),
                ])

X_train_final = all_pipeline.fit_transform(X_train)
X_test_final = all_pipeline.transform(X_test)
```

```python
y_train.value_counts()
```

```python
X_train_final.shape, X_test_final.shape
```

```python
# Dump the pipeline locally in ARTIFACTS_FOLDER_PATH
ARTIFACTS_FOLDER_PATH = os.path.join(os.getcwd(), '..', 'artifacts')
os.makedirs(ARTIFACTS_FOLDER_PATH, exist_ok=True)
```

```python
# Dump the pipeline locally --> Serialzation
joblib.dump(all_pipeline, os.path.join(ARTIFACTS_FOLDER_PATH, 'pipeline.pkl'))
```

```python
# Transform X_train and X_test
X_train_final = all_pipeline.fit_transform(X_train)
X_test_final = all_pipeline.transform(X_test)

# Fill NaN values in y_train and y_test with their medians
y_train_filled = y_train.fillna(y_train.median())
y_test_filled = y_test.fillna(y_test.median())
```

# `Building ML Models`

evaluation metric : RMSE

```python
#train
lin_reg = LinearRegression()
lin_reg.fit(X_train_final, y_train_filled)
#pre
y_pred_train_lin = lin_reg.predict(X_train_final)
y_pred_test_lin = lin_reg.predict(X_test_final)
# how much isi true 
rmse_train_lin = np.sqrt(mean_squared_error(y_train_filled, y_pred_train_lin))
rmse_test_lin = np.sqrt(mean_squared_error(y_test_filled, y_pred_test_lin))

print(f'RMSE for Training using LinearRegression is: {rmse_train_lin:.3f}')
print(f'RMSE for Testing using LinearRegression is: {rmse_test_lin:.3f}')
```

```python
# is more perfction pecause is meansqure is knn leser than linear reg

knn_reg = KNeighborsRegressor()
knn_reg.fit(X_train_final, y_train_filled)

y_pred_train_knn = knn_reg.predict(X_train_final)
y_pred_test_knn = knn_reg.predict(X_test_final)

rmse_train_knn = np.sqrt(mean_squared_error(y_train_filled, y_pred_train_lin))
rmse_test_knn = np.sqrt(mean_squared_error(y_test_filled, y_pred_test_knn))

print(f'RMSE for Training using KNeighborsRegressor is: {rmse_train_knn:.3f}')
print(f'RMSE for Testing using KNeighborsRegressor is: {rmse_test_knn:.3f}')
```

```python
forest_reg = RandomForestRegressor(n_estimators=500, max_depth=30, random_state=45)
forest_reg.fit(X_train_final,y_train_filled)

# Predict and evaluate
y_pred_train_forest = forest_reg.predict(X_train_final)
y_pred_test_foerst = forest_reg.predict(X_test_final)

# Using fRMSE
RMSE_train_forest = np.sqrt(mean_squared_error(y_train_filled, y_pred_train_forest))
RMSE_test_forest = np.sqrt(mean_squared_error(y_test_filled, y_pred_test_foerst))

print(f'The RMSE using root mean squared error model for Training Dataset  {RMSE_train_forest:.3f}')
print(f'The RMSE using root mean squared error model for Testing Dataset  {RMSE_test_forest:.3f}')
```

```python
svm_model = SVR(kernel='linear')  # You can change the kernel to 'linear' or 'poly' as needed
svm_model.fit(X_train_final, y_train_filled)

# Predict
y_pred_train_svm = svm_model.predict(X_train_final)
y_pred_test_svm = svm_model.predict(X_test_final)

# Evaluate
rmse_train_svm = np.sqrt(mean_squared_error(y_train_filled, y_pred_train_svm))
rmse_test_svm = np.sqrt(mean_squared_error(y_test_filled, y_pred_test_svm))

print(f'RMSE for Training using SVR is: {rmse_train_svm:.3f}')
print(f'RMSE for Testing using SVR is: {rmse_test_svm:.3f}')
```

```python
joblib.dump(lin_reg, os.path.join(ARTIFACTS_FOLDER_PATH, 'lin_reg.pkl'), compress=9)
joblib.dump(forest_reg, os.path.join(ARTIFACTS_FOLDER_PATH, 'forest_reg.pkl'), compress=9)
joblib.dump(knn_reg, os.path.join(ARTIFACTS_FOLDER_PATH, 'knn_reg.pkl'), compress=9)
joblib.dump(svm_model , os.path.join(ARTIFACTS_FOLDER_PATH, 'svm_model.pkl'), compress=9)
```
